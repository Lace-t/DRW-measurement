```python
import numpy as np
import matplotlib.pyplot as plt

c1="#009ad6"
c2='#f15a22'
c3='#7fb80e'
c4='#6f599c'
```

## N=60 y[0]=m

```python
data1=np.loadtxt('s60m.txt')
data1.shape
```

```python
tau=data1[:,1:10001]/data1[:,0].reshape(61,1)
sigma=data1[:,10001:]/0.14
tin=2922/data1[:,0]
```

```python
plt.figure(dpi=150)
plt.plot(tin,np.ones(len(tin)),'k')

plt.plot(tin,np.mean(tau,axis=1),color=c1,label=r'mean(S21)')
plt.plot(tin,np.mean(tau,axis=1)+np.std(tau,axis=1),'--',color=c1,label=r'$\sigma_G$(S21)')
plt.plot(tin,np.mean(tau,axis=1)-np.std(tau,axis=1),'--',color=c1)

plt.xscale('log')
plt.xlabel(r'$T/\tau_{in}$',fontsize=15)
plt.ylabel(r'$\tau_{out}/\tau_{in}$',fontsize=15)
plt.title(r"N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS},y[0]=m$",fontsize=20)
#plt.title(r"N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0,2)
plt.legend(fontsize=15,loc=4)
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(tin,np.ones(len(tin)),'k')

plt.plot(tin,np.median(tau,axis=1),color=c1,label=r'median(S21)')
plt.plot(tin,np.percentile(tau,16,axis=1),'--',color=c1,label=r'$\sigma_G$(S21)')
plt.plot(tin,np.percentile(tau,84,axis=1),'--',color=c1)

plt.xscale('log')
plt.xlabel(r'$T/\tau_{in}$',fontsize=15)
plt.ylabel(r'$\tau_{out}/\tau_{in}$',fontsize=15)
plt.title(r"N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS},y[0]=m$",fontsize=20)
#plt.title(r"N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0,2)
plt.legend(fontsize=15,loc=4)
plt.show()
```

### 双百检验

```python
plt.figure(dpi=150)
plt.plot(tin,np.ones(len(tin)),'k')

for i in range(100):
    hundred=np.array(list(map(lambda x: np.random.choice(x,100),tau)))
    if i==0:
        plt.plot(tin,np.median(hundred,axis=1),color=c1,label=r'median(100)',alpha=0.05)
    else:
        plt.plot(tin,np.median(hundred,axis=1),color=c1,alpha=0.05)
plt.plot(tin,np.median(tau,axis=1),color=c1,label=r'median($10^4$)')

plt.xscale('log')
plt.xlabel(r'$T/\tau_{in}$',fontsize=15)
plt.ylabel(r'$\tau_{out}/\tau_{in}$',fontsize=15)
plt.title(r"N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS},y[0]=m$",fontsize=20)
#plt.title(r"N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0.25,1.5)
plt.legend(fontsize=15,loc=4)
plt.show()
```

```python
plt.figure(dpi=150)

for i in range(100):
    hundred=np.array(list(map(lambda x: np.random.choice(x,100),data1[:,1:10001])))
    if i==0:
        plt.plot(-np.log10(tin),np.log10(np.median(hundred,axis=1)/2922),color=c1,label=r'median(100)',alpha=0.1)
    else:
        plt.plot(-np.log10(tin),np.log10(np.median(hundred,axis=1)/2922),color=c1,alpha=1)
plt.plot(-np.log10(tin),np.log10(np.median(data1[:,1:10001],axis=1)/2922),color='r',label=r'median($10^4$)')
plt.plot(-np.log10(tin),-np.log10(tin),'k')
#plt.xscale('log')
#plt.yscale('log')
plt.xlabel(r'$\lg(\tau_{in}/T)$',fontsize=15)
plt.ylabel(r'$\lg(\tau_{out}/T)$',fontsize=15)
plt.title(r"N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS},y[0]=m$",fontsize=20)
#plt.title(r"N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(0,-3)
plt.grid()
plt.ylim(0,-3)
plt.legend(fontsize=15,loc=4)
plt.show()
```

```python
data=np.loadtxt('s445m.txt')
data2=np.loadtxt('s445.txt')
x=np.load('de445j1.npy')
```

```python
tm=data[:,1:10001]/data[:,0].reshape(61,1)
tau=data2[:,1:10001]/data2[:,0].reshape(61,1)
tin=2922/data[:,0]
pe=x[:,:,2]/data[:,0].reshape(61,1)
```

```python
plt.figure(dpi=150)
plt.plot(tin,np.ones(len(tin)),'k:',lw=2)

for i in range(100):
    hundred=np.array(list(map(lambda x: np.random.choice(x,100),tm)))
    if i==0:
        plt.plot(tin,np.median(hundred,axis=1),color='0.4',lw=1,label=r'S21: fixed + grid + $10^2$ LCs',
                alpha=0.1)
    else:
        plt.plot(tin,np.median(hundred,axis=1),color='0.4',lw=1,alpha=0.1)
plt.plot(tin,np.median(tm,axis=1),color='k',lw=2,label=r'S21: fixed + grid + $10^4$ LCs')
plt.plot(tin,np.median(tau,axis=1),color=c2,lw=2,label=r'S21: free + grid + $10^4$ LCs')
plt.plot(tin,np.median(pe,axis=1),'--',lw=4,color=c1,label=r'Ours: free + MCMC + $10^4$ LCs')

plt.xscale('log')
plt.xlabel(r'$T/\tau_{\rm in}$',fontsize=20)
plt.ylabel(r'$\tau_{\rm out}^{\rm median}/\tau_{\rm in}$',fontsize=20)
#plt.title(r"N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS},y[0]=m$",fontsize=20)
plt.title(r"S21 prior (N = 445, $\sigma_{\rm e}=\sigma_{\rm OGLE}$)",fontsize=20)
plt.text(10,1.2,r'S21PE50: median of $\tau_{\rm PE}$',fontsize=18)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0.3,1.3)
plt.legend(fontsize=14,loc=4)
plt.show()
```
