```python
import numpy as np
import matplotlib.pyplot as plt

c1="#009ad6" #PE
c2='#f15a22' #MLE
c3='#7fb80e' #PM
c4='#aa2116' #MAP
c5='#ffe600' #
rho=np.logspace(3,0,61)
tin=np.logspace(0.4657,3.4657,61)
```

```python
x=np.load('de60j2.npy')
x.shape
```

```python
tau=x[:,:,0:4]/tin.reshape(61,1,1)
sigma=x[:,:,4:]/0.14
```

```python
plt.figure(dpi=150)
plt.plot(rho,np.ones(len(rho)),'k')

plt.plot(rho,np.mean(tau,axis=1)[:,2],'--',color=c1,lw=5,label=r'$\tau_{\rm PE}$')
plt.plot(rho,np.mean(tau,axis=1)[:,3],color=c3,lw=5,label=r'$\tau_{\rm PM}$')
plt.plot(rho,np.mean(tau,axis=1)[:,1],'-.',color=c4,lw=5,label=r'$\tau_{\rm MAP}$')
plt.plot(rho,np.mean(tau,axis=1)[:,0],':',color=c2,lw=5,label=r'$\tau_{\rm MLE}$')

plt.xscale('log')
plt.xlabel(r'$T/\tau_{\rm in}$',fontsize=20)
plt.ylabel(r'$\tau_{\rm out}^{\rm mean}/\tau_{\rm in}$',fontsize=20)
plt.text(10,0.1,'mean',fontsize=20)
plt.title(r'K17 prior(N=60,$\sigma_e=\sigma_{\rm SDSS}$)',fontsize=20)
#plt.title(r'S21 prior(N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$)',fontsize=20)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0,1.5)
plt.legend(fontsize=15,loc=4)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(rho,np.ones(len(rho)),'k')

plt.plot(rho,np.median(tau,axis=1)[:,2],'--',color=c1,lw=5,label=r'$\tau_{\rm PE}$')
plt.plot(rho,np.median(tau,axis=1)[:,3],color=c3,lw=5,label=r'$\tau_{\rm PM}$')
plt.plot(rho,np.median(tau,axis=1)[:,1],'-.',color=c4,lw=5,label=r'$\tau_{\rm MAP}$')
plt.plot(rho,np.median(tau,axis=1)[:,0],':',color=c2,lw=5,label=r'$\tau_{\rm MLE}$')

plt.xscale('log')
plt.xlabel(r'$T/\tau_{\rm in}$',fontsize=20)
plt.ylabel(r'$\tau_{\rm out}^{\rm median}/\tau_{\rm in}$',fontsize=20)
plt.text(10,0.1,'median',fontsize=20)
plt.title(r'K17 prior(N=60,$\sigma_e=\sigma_{\rm SDSS}$)',fontsize=20)
#plt.title(r'S21 prior(N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$)',fontsize=20)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0,1.5)
plt.legend(fontsize=15,loc=4)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
plt.figure(dpi=150)
plt.plot(rho,np.ones(len(rho)),'k')
pm=tau[:,:,3]
for i in range(100):
    hundred=np.array(list(map(lambda x: np.random.choice(x,100),pm)))
    if i==0:
        plt.plot(rho,np.mean(hundred,axis=1),color=c3,label=r'mean(100)',alpha=0.05)
    else:
        plt.plot(rho,np.mean(hundred,axis=1),color=c3,alpha=0.05)
plt.plot(rho,np.mean(pm,axis=1),color=c3,label=r'mean($10^4$)')

plt.xscale('log')
plt.xlabel(r'$T/\tau_{in}$',fontsize=15)
plt.ylabel(r'$\tau_{out}/\tau_{in}$',fontsize=15)
plt.title(r"K17 prior N=60,$\sigma_{\rm e}=\sigma_{\rm SDSS}$",fontsize=20)
#plt.title(r"N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$",fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0.25,1.5)
plt.legend(fontsize=15,loc=4)
plt.show()
```

## 双百检验

```python
x1=np.load('de60j1.npy')
x2=np.load('de60j2.npy')
#[0]MLE [1]MAP [2]PE [3]PM
```

```python
x1[:,:,2].shape
```

```python
tau1=x1[:,:,2]/tin.reshape(61,1)
tau2=x2[:,:,3]/tin.reshape(61,1)
```

```python
tau1.shape
```

```python
mse1=np.array([])
mse2=np.array([])
for i in range(100):
    hundred1=np.array(list(map(lambda x: np.random.choice(x,100),tau1)))
    hundred2=np.array(list(map(lambda x: np.random.choice(x,100),tau2)))
    mse1=np.append(mse1,np.power(1-np.median(hundred1,axis=1),2))
    mse2=np.append(mse2,np.power(1-np.mean(hundred2,axis=1),2))
mse1.resize(100,61);mse2.resize(100,61)
mse1=mse1.T;mse2=mse2.T
```

```python
plt.figure(dpi=150)
plt.plot(rho,np.mean(mse1,axis=1),color=c1,label=r'S21 prior $\tau_{PE}$ median(100)')
plt.fill_between(rho,np.mean(mse1,axis=1)+np.std(mse1,axis=1)/10,np.mean(mse1,axis=1)-np.std(mse1,axis=1)/10,
                 color=c1,alpha=0.3,edgecolor="none")
plt.plot(rho,np.mean(mse2,axis=1),color=c3,label=r'K17 prior $\tau_{PM}$ mean(100)')
plt.fill_between(rho,np.mean(mse2,axis=1)+np.std(mse2,axis=1)/10,np.mean(mse2,axis=1)-np.std(mse2,axis=1)/10,
                 color=c3,alpha=0.3,edgecolor="none")
plt.xscale('log')
plt.yscale('log')
plt.xlabel(r'$T/\tau_{\rm in}$',fontsize=20)
plt.ylabel(r'MSE',fontsize=20)
plt.title(r'N=60,$\sigma_e=\sigma_{\rm SDSS}$',fontsize=20)
#plt.title(r'S21 prior(N=445,$\sigma_{\rm e}=\sigma_{\rm OGLE}$)',fontsize=20)
plt.xlim(1,1000)
plt.ylim(1e-4,1)
plt.grid()
plt.legend(fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
