```python
import numpy as np
import matplotlib.pyplot as plt

c1="#009ad6"
c2='#f15a22'
c3='#7fb80e'
c4='#aa2116'
c5='#ffe600'
rho=np.logspace(3,0,61)
```

```python
data=np.loadtxt('k445.txt')
```

```python
k17=np.median(data,axis=1)
```

```python
x=np.loadtxt('de445j2.txt')
```

```python
k17=k17/x[:,0]
MAP=x[:,6]/x[:,0]
```

```python
plt.figure(dpi=150)
plt.plot(rho,np.ones(len(rho)),'k')

plt.plot(rho,k17,color=c2,lw=2,label=r'K17: PRH')
plt.plot(rho,MAP,'-.',color=c4,lw=3,label=r'Ours: CELERITE')


plt.xscale('log')
plt.xlabel(r'$T/\tau_{\rm in}$',fontsize=20)
plt.ylabel(r'$\tau_{\rm out}^{\rm median}/\tau_{\rm in}$',fontsize=20)
plt.text(8,1.2,r'K17MAP50: median of $\tau_{\rm MAP}$',fontsize=18)
#plt.title(r'K17 prior (N = 60, $\sigma_e=\sigma_{\rm SDSS})$',fontsize=20)
plt.title(r'K17 prior (N=445, $\sigma_{\rm e}=\sigma_{\rm OGLE})$',fontsize=20)
plt.xlim(1,1000)
plt.grid()
plt.ylim(0.1,1.3)
plt.legend(fontsize=14,loc=4)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
